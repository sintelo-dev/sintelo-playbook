# 05 – Evaluation  
## AdventureWorks | Capital Operativo

### Propósito de la fase

Evaluar las señales generadas por el modelo para entender patrones,
concentraciones y desalineaciones relevantes entre la asignación del
capital operativo (inventario) y la generación de ventas.

Esta fase no busca tomar decisiones ni emitir recomendaciones, sino
habilitar orientación estratégica con base empírica.

---

## Alcance de la evaluación

La evaluación se realiza a dos niveles de granularidad:

1. **Categoría de producto**
2. **Subcategoría de producto**

El objetivo es verificar:
- La consistencia de las señales observadas
- Si las desalineaciones son artefactos de agregación o patrones estructurales

---

## Evaluación a nivel Categoría

### Observación 1 — Concentración de capital operativo

La asignación del capital operativo se encuentra altamente concentrada en
una sola categoría (Bikes), que representa la mayor proporción del
inventario total del sistema.

Esta concentración define la estructura base sobre la cual se distribuye
el resto del capital operativo.

---

### Observación 2 — Concentración de generación de ventas

La generación de ventas también se concentra mayoritariamente en la misma
categoría dominante (Bikes), aunque en una proporción distinta a la del
capital asignado.

Esto indica que la concentración de ventas y la concentración de capital
no son perfectamente proporcionales.

---

### Observación 3 — Desalineaciones relativas entre capital y ventas

Al comparar la distribución relativa del capital operativo con la
distribución relativa de ventas, se observan brechas (alignment gaps):

- Categorías que generan una proporción mayor de ventas que el capital que
  consumen.
- Categorías que consumen una proporción mayor de capital que la proporción
  de ventas que generan.

A nivel categoría, la desalineación negativa más relevante se concentra en
Components.

---

## Evaluación a nivel Subcategoría

### Observación 4 — Validación de granularidad

El análisis a nivel subcategoría confirma que las señales observadas a
nivel categoría no desaparecen al aumentar la granularidad del modelo.

Las desalineaciones identificadas se mantienen consistentes, lo que
indica que no son producto de agregación excesiva.

---

### Observación 5 — Comportamiento heterogéneo dentro de Bikes

Dentro de la categoría Bikes, el desempeño no es homogéneo entre
subcategorías:

- Algunas subcategorías presentan gaps positivos relevantes.
- Otras subcategorías muestran desalineaciones negativas moderadas.

Esto sugiere que el gap positivo observado a nivel categoría se explica
principalmente por subcategorías específicas y no por un desempeño
uniforme del portafolio completo.

---

### Observación 6 — Desalineación estructural en Components

En la categoría Components, la desalineación negativa no se explica por
una sola subcategoría aislada.

Por el contrario, se observa un patrón distribuido de subcategorías que:
- Consumen una proporción relevante del capital operativo
- Generan una proporción menor de ventas relativas

Esto indica una desalineación de carácter estructural dentro de la
categoría.

---

### Observación 7 — Categorías de baja materialidad sistémica

Las categorías Clothing y Accessories presentan múltiples subcategorías
con gaps positivos.

Sin embargo, su impacto sistémico es limitado debido a su bajo peso
relativo en el capital operativo total.

Estas categorías no explican los desbalances principales del sistema, aun
cuando muestran eficiencia relativa.

---

## Implicaciones de la evaluación

El modelo permite afirmar que:

- La asignación del capital operativo no es proporcional a la generación
  de ventas en todas las categorías.
- Las desalineaciones observadas son consistentes al analizar distintos
  niveles de granularidad.
- Existen patrones estructurales que no pueden identificarse sin este tipo
  de comparación relativa.

Estas observaciones eliminan la posibilidad de discutir la asignación de
capital operativo sin referencia empírica.

---

## Fuera de alcance de esta fase

Esta evaluación no:

- Determina causas operativas o estratégicas
- Evalúa rentabilidad o margen
- Recomienda acciones o cambios de política
- Propone optimización de inventarios

Su función es exclusivamente orientar la discusión estratégica.

---

## Cierre — Evaluation

La fase de evaluación confirma que el modelo desarrollado proporciona una
lectura clara y consistente de la relación entre capital operativo y
generación de ventas.

Las señales identificadas no constituyen decisiones, pero habilitan
conversaciones estratégicas que previamente no podían sostenerse con base
en datos estructurados.

El siguiente paso natural sería definir qué decisiones estratégicas
requieren mayor profundidad analítica y qué información adicional sería
necesaria para habilitarlas.
