# 01 – Business Understanding  
## AdventureWorks | Capital Operativo

---

## 1. Contexto del negocio

AdventureWorks opera como una empresa de manufactura discreta con una operación intensiva en inventarios.  
Una porción relevante de su capital operativo se encuentra inmovilizada en:

- Inventarios de producto terminado  
- Inventarios en proceso  
- Inventarios de componentes  

Desde una perspectiva financiera y operativa, el inventario representa una decisión de asignación de capital.  
Antes de considerar cualquier acción correctiva, la organización requiere claridad sobre cómo está distribuido ese capital y qué retorno observable está generando.

Este proyecto busca habilitar esa claridad inicial.

---

## 2. Problema de negocio

La pregunta que guía el proyecto es:

> ¿Cómo se distribuye el capital operativo de AdventureWorks entre inventario,  
> y qué tan alineada está esa distribución con la generación de ventas?

El análisis no parte de la premisa de ineficiencia.  
Su objetivo es evitar decisiones basadas en intuición o métricas aisladas.

---

## 3. Decisión que se busca habilitar

Este análisis no toma decisiones operativas ni financieras.  
Su propósito es habilitar la orientación necesaria para responder preguntas como:

- ¿En qué segmentos del inventario está concentrado el capital?
- ¿Qué segmentos generan ventas proporcionales a ese capital?
- ¿Dónde podrían existir desalineaciones relevantes?

Las decisiones posteriores quedan fuera de este alcance.

---

## 4. Objetivo del análisis

El objetivo es construir un primer diagnóstico claro, defendible y reproducible sobre la alineación entre:

- Uso del capital operativo (inventario)
- Generación de ventas (retorno observable)

El diagnóstico debe permitir comparaciones relativas y la formulación de preguntas de decisión.

---

## 5. Alcance del proyecto

### Dentro del alcance

- Análisis descriptivo
- Segmentación de inventario
- Métricas relativas (porcentajes)
- Comparación entre uso de capital y ventas
- Evidencia trazable y reproducible

### Fuera del alcance

- Optimización de inventarios
- Modelos predictivos
- Recomendaciones tácticas
- Cambios de política operativa
- Simulaciones o escenarios

Este proyecto termina en claridad, no en acción.

---

## 6. Definiciones de alto nivel

### Capital operativo

Capital inmovilizado en inventario, medido mediante proxies contables y operativos.

### Retorno

Generación de ventas observable asociada a los productos en inventario.

No se analizan márgenes, rentabilidad, flujo de caja ni costo de capital.

---

## 7. Entregable esperado

El entregable técnico central es una tabla analítica que permita observar, por segmento:

- Porcentaje de inventario
- Porcentaje de capital operativo
- Porcentaje de ventas

La tabla permitirá identificar:

- Segmentos con alto consumo de capital y bajo aporte a ventas
- Segmentos alineados
- Segmentos potencialmente sobreinvertidos

Esta tabla no es el deployment final, sino el insumo para un memo de decisión posterior.

---

## 8. Rol del equipo de Ciencia de Datos

El equipo de Ciencia de Datos es responsable de:

1. Construir la tabla analítica de forma correcta y consistente  
2. Documentar supuestos y definiciones  
3. Asegurar reproducibilidad  
4. Preparar el terreno para interpretación ejecutiva  

El equipo no toma decisiones, no recomienda acciones y no optimiza.

---

## 9. Criterios de éxito

El análisis se considera exitoso si:

- Un CFO u operador puede entenderlo sin contexto técnico
- Las métricas están claramente definidas
- La tabla permite formular preguntas de decisión
- No existe ambigüedad sobre qué se mide y qué no

---

## 10. Relación con la metodología

### OODA

- Observe: datos operativos y financieros  
- Orient: comparación entre inventario, capital y ventas  
- Decide / Act: fuera de alcance  

### CRISP-DM

- Business Understanding  
- Data Understanding  
- Data Preparation (siguiente etapa)  
- Analysis (descriptivo)  
- Deployment (fuera de alcance)

---

## 11. Frase guía

Este análisis no decide, pero hace imposible seguir decidiendo a ciegas.
